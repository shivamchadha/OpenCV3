```python
from matplotlib import pyplot
import cv2 
import numpy as np 
```

```python

img = cv2.imread(r"C:\Users\shiva\OneDrive\Pictures\Screenshots\tobi__the_masked_ninja_from_akatsuki_by_beu2v_dbtmcz6.jpg", cv2.IMREAD_COLOR) 
  
 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) 
  
# Blur 
#gray_blurred = cv2.blur(gray, (3, 3)) 
  

detected_circles = cv2.HoughCircles(gray,  cv2.HOUGH_GRADIENT,1,50 ,10,param1 = 300,param2 = 36, minRadius = 15, maxRadius = 110) 
  
 
if detected_circles is not None: 
  
     
    detected_circles = np.uint16(np.around(detected_circles))  
  
    for pt in detected_circles[0, :]: 
        a, b, r = pt[0], pt[1], pt[2] 
  
         
        cv2.circle(img, (a, b), r, (0, 255, 0), 2) 
  
        #to show the center. 
        #cv2.circle(img, (a, b), 1, (0, 0, 255), 3) 
        pyplot.imshow(img)
        
        #cv2.waitKey(0) 
```
